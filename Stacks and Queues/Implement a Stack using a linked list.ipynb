```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class Stack:
    def __init__(self):
        self.head = None
        self.num_of_items = 0 
        
    def push(self, value):
        if self.head == None:
            self.head = Node(value)
        else:
            new_node = Node(value)
            new_node.next = self.head
            self.head = new_node
        self.num_of_items += 1
    
    def size(self):
#         current = self.head
#         count = 0
#         while(current):
#             count += 1
#             current = current.next
        return self.num_of_items

    def pop(self):
        if not self.is_empty():
            item = self.head
            self.head = self.head.next
            self.num_of_items -= 1
            return item.value
        return None
    
    def is_empty(self):
        return self.num_of_items == 0
```
