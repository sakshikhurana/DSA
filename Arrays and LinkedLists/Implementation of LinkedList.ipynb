Write the code for the create_linked_list function below
The function should take a Python list of values as input and return the head of a linked list that has those values

```python
def create_linked_list(py_list):
    try:
        head = Node(py_list[0])
        current_node = head
        for i in range(1,len(py_list)):
            current_node.next = Node(py_list[i])
            current_node = current_node.next
    except IndexError:
        head = None
    return head
```

```python
link_list = create_linked_list([1,2,3,4])
```
