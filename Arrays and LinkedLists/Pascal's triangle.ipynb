# Problem Statement
Find and return the nth row of Pascal's triangle in the form a list. n is 0-based.

For exmaple, if n = 4, then output = [1, 4, 6, 4, 1].

To know more about Pascal's triangle: https://www.mathsisfun.com/pascals-triangle.html

```python
def factorial(num):
    """Finds factorial of a number"""
    if num == 0:
        return 1
    return num * factorial(num-1)

def combinations(n, k):
    """Combinations of n!/(n-k)!k!"""
    return int(factorial(n)/(factorial(n-k)*factorial(k)))

def pascals_triangle(n):
    """List of row of Pascal's triangle"""
    final_result = []
    for k in range(n+1):
        num = combinations(n,k)
        final_result.append(num)
    print(final_result)
```

```python
pascals_triangle(4)
```
