# Problem Statement
Given a linked list, swap the two nodes present at position i and j. The positions are based on 0-based indexing.

Note: You have to swap the nodes and not just the values.

Example:

linked_list = 3 4 5 2 6 1 9
positions = 3 4
output = 3 4 5 6 2 1 9


```python
class Node:
    def __init__(self, data):
        self.value = data
        self.next = None

def create_linked_list(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    tail = head
    for data in arr[1:]:
        tail.next = Node(data)
        tail = tail.next
    return head

def print_linked_list(head):
    while head:
        print(head.value, end=' ')
        head = head.next
    print()

def swap_nodes(head, pos1, pos2):
    current = head
    index = 0
    prev1 = node1 = prev2 =  node2 = None
    if pos1 == pos2:
      return head
    if index == pos2:
      node2 = current
    if index == pos1:
        node1 = current
    while current:
        if index+1 == pos1:
            prev1 = current
            node1 = current.next
        if index+1 == pos2:
            prev2 = current
            node2 = current.next
        if node1 and node2:
            if prev1 == None:
                head = node2
            else:
                prev1.next = node2
            if prev2 == None:
                head = node1
            else:
                prev2.next = node1
        
            tmp = node1.next
            node1.next = node2.next
            node2.next = tmp
            return head
        current = current.next
        index +=1
    return head
    
```
