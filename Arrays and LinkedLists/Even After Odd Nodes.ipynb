# Problem Statement
Given a linked list with integer data, arrange the elements in such a manner that all nodes with even numbers are placed after odd numbers. Do not create any new nodes and avoid using any other data structure. The relative order of even and odd elements must not change.

Example:

linked list = 1 2 3 4 5 6
output = 1 3 5 2 4 6

```python
class Node:
    def __init__(self, value):
        self.next = None
        self.value = value
```

```python
def even_odd(head):
    last_even = even_head = last_odd = odd_head = None
    current = head
    while current:
        if current.value % 2 == 0:
            if not even_head:
                even_head = last_even = current
            else:
                last_even.next = current
                last_even = current
        else:
            if not odd_head:
                odd_head = last_odd = current
            else:
                last_odd.next = current
                last_odd = current
        current = current.next
    if odd_head != None:
        head = odd_head
    if last_odd != None:
        last_odd.next = even_head
    if last_even != None:
        last_even.next = None
    
    return head
                
```

```python
def create_linked_list(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    tail = head
    for data in arr[1:]:
        tail.next = Node(data)
        tail = tail.next
    return head

def print_linked_list(head):
    while head:
        print(head.value, end=' ')
        head = head.next
    print()
```

```python
arr = [0, 4, 1, 5 ,6]
head = create_linked_list(arr)
```

```python
linked_list = even_odd(head)
```

```python
print_linked_list(linked_list)
```
